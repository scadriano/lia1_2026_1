## 🏗️ Criando Arrays

### 📋 Principais métodos de criação:
- **`np.array()`** - A partir de listas/tuplas
- **`np.arange()`** - Sequências numéricas
- **`np.linspace()`** - Valores igualmente espaçados
- **`np.zeros()`** - Arrays preenchidos com zeros
- **`np.ones()`** - Arrays preenchidos com uns
- **`np.random.random()`** - Arrays com valores aleatórios

````python
# Importando a biblioteca NumPy
import numpy as np
print(f"NumPy versão: {np.__version__}")
````

````python
# A partir de uma lista Python
lista = [1, 2, 3, 4, 5]
array_de_lista = np.array(lista)
print(f"Array a partir de lista: {array_de_lista}")

````

````python
# Criando um array com uma sequência de valores (de 0 a 8, pulando de 2 em 2)
array_arange = np.arange(0, 10, 2)
print(f"Array com arange: {array_arange}")
````

````python
# Criando um array de 5 números igualmente espaçados entre 0 e 1
array_linspace = np.linspace(0, 1, 5)
print(f"Array com linspace: {array_linspace}")
````

````python
# Criando uma matriz (array 2D) de zeros
matriz_zeros = np.zeros((2, 3))
print("Matriz de zeros:\n", matriz_zeros)


````

````python
# Criando uma matriz (array 2D) de uns
matriz_uns = np.ones((3, 4))
print("Matriz de uns:")
print(matriz_uns)
````

````python
# Criando uma matriz identidade
matriz_identidade = np.eye(3)
print("\nMatriz identidade:")
print(matriz_identidade)
````

````python
# Criando um array com valores aleatórios
array_aleatorio = np.random.random((2, 3))
print("\nArray aleatório:")
print(array_aleatorio)
````

### 🔍 Propriedades dos Arrays


````python
# Explorando propriedades dos arrays
matriz = np.array([[1, 2, 3], [4, 5, 6]])

print("Array original:")
print(matriz)
print(f"Forma (shape): {matriz.shape}")
print(f"Dimensões (ndim): {matriz.ndim}")
print(f"Tamanho total (size): {matriz.size}")
print(f"Tipo de dados (dtype): {matriz.dtype}")
print(f"Bytes por elemento: {matriz.itemsize}")
print(f"Memória total: {matriz.nbytes} bytes")
````

## Operações com Arrays

````python
# Mais operações matemáticas
x = np.array([1, 4, 9, 16, 25])
y = np.array([2, 3, 4, 5, 6])

print("Array x:", x)
print("Array y:", y)
print()

# Raiz quadrada
print("Raiz quadrada de x:", np.sqrt(x))

# Potência
print("y elevado ao quadrado:", np.power(y, 2))

# Funções trigonométricas
angulos = np.array([0, np.pi/4, np.pi/2, np.pi])
print("Seno dos ângulos:", np.sin(angulos))

# Exponencial e logaritmo
print("Exponencial de [1, 2, 3]:", np.exp([1, 2, 3]))
print("Logaritmo natural de [1, 2, 3]:", np.log([1, 2, 3]))
````

````python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

soma = a + b
print(f"Soma: {soma}") 

multiplicacao = a * 2
print(f"Multiplicação por escalar: {multiplicacao}") 

# Broadcasting: operando com arrays de tamanhos diferentes
matriz = np.array([[1, 2, 3], [4, 5, 6]])
escalar = 10

soma_broadcast = matriz + escalar
print("Soma com broadcasting:\n", soma_broadcast)
````

### 📊 Funções estatísticas e agregações


````python
# Criando dados para análise estatística
np.random.seed(42)  # Para reproduzibilidade
dados_vendas = np.random.normal(1000, 200, 100)  # 100 vendas com média 1000 e desvio 200

print("Dados de vendas (primeiros 10 valores):")
print(dados_vendas[:10])
print()

# Estatísticas descritivas
print("=== ESTATÍSTICAS DESCRITIVAS ===")
print(f"Média: {np.mean(dados_vendas):.2f}")
print(f"Mediana: {np.median(dados_vendas):.2f}")
print(f"Desvio padrão: {np.std(dados_vendas):.2f}")
print(f"Variância: {np.var(dados_vendas):.2f}")
print(f"Mínimo: {np.min(dados_vendas):.2f}")
print(f"Máximo: {np.max(dados_vendas):.2f}")
print()

# Percentis
print("=== PERCENTIS ===")
print(f"25º percentil: {np.percentile(dados_vendas, 25):.2f}")
print(f"50º percentil (mediana): {np.percentile(dados_vendas, 50):.2f}")
print(f"75º percentil: {np.percentile(dados_vendas, 75):.2f}")
print()

# Soma e produto
print("=== AGREGAÇÕES ===")
print(f"Soma total: {np.sum(dados_vendas):.2f}")
print(f"Produto dos 5 primeiros: {np.prod(dados_vendas[:5]):.2e}")
print(f"Soma acumulada (primeiros 5): {np.cumsum(dados_vendas[:5])}")
````

### 📐 Manipulação de Formas e Dimensões


````python
# Criando um array para manipulação
arr = np.arange(12)
print("Array original:")
print(f"Valores: {arr}")
print(f"Forma: {arr.shape}")
print()

# Reshape - mudando a forma
arr_2d = arr.reshape(3, 4)
print("Array 2D (3x4):")
print(arr_2d)
print(f"Forma: {arr_2d.shape}")
print()

# Reshape para 3D
arr_3d = arr.reshape(2, 2, 3)
print("Array 3D (2x2x3):")
print(arr_3d)
print(f"Forma: {arr_3d.shape}")
print()

# Flatten - achatar o array
arr_flat = arr_3d.flatten()
print("Array achatado:")
print(f"Valores: {arr_flat}")
print(f"Forma: {arr_flat.shape}")
print()


````

````python
# Transpose - transpor matriz
matriz = np.array([[1, 2, 3], [4, 5, 6]])
print("Matriz original:")
print(matriz)
print(f"Forma: {matriz.shape}")
print()

matriz_t = matriz.T
print("Matriz transposta:")
print(matriz_t)
print(f"Forma: {matriz_t.shape}")
print()

# Concatenação
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
concat_h = np.concatenate([arr1, arr2])
print(f"Concatenação horizontal: {concat_h}")

# Stack vertical
stack_v = np.vstack([arr1, arr2])
print("Stack vertical:")
print(stack_v)
````

## Indexação e Slicing

````python
# Array para os exemplos
dados = np.arange(10, 20)
print(f"Array original: {dados}")
````

````python
# Acessando um elemento (o terceiro elemento, índice 2)
print(f"Elemento no índice 2: {dados[2]}")
````

````python
# Slicing: pegando um subconjunto (do índice 2 ao 5)
print(f"Slicing de 2 a 5: {dados[2:6]}")
````

````python
# Indexação booleana: pegando apenas os números maiores que 15
maiores_que_15 = dados[dados > 15]
print(f"Números maiores que 15: {maiores_que_15}")

# Indexação booleana com operações lógicas
maiores_que_15_ou_menores_que_12 = dados[(dados > 15) | (dados < 12)]
print(f"Números maiores que 15 ou menores que 12: {maiores_que_15_ou_menores_que_12}")
````

````python
# Usando np.where para substituir valores
# Substitui números pares por -1
dados_modificados = np.where(dados % 2 == 0, -1, dados)
print(f"Pares substituídos por -1: {dados_modificados}")
````

# 📊 Pandas: Análise e Manipulação de Dados

**O que é Pandas? Series e DataFrames**

- Pandas é construído sobre o NumPy e facilita a manipulação de dados tabulares.   

- As duas estruturas principais são:   

    - Series: Um array unidimensional com rótulos (índices). Pense nela como uma única coluna de uma tabela.   

    - DataFrame: Uma estrutura bidimensional, como uma tabela ou planilha, onde cada coluna é uma Series.   



## Criando Series e DataFrames

````python
# Importando a biblioteca Pandas
import pandas as pd
print(f"Pandas versão: {pd.__version__}")

# Criando uma Series a partir de uma lista
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
print("Series:")
print(s)

````

## 🔄 Transformação e Limpeza de Dados

Uma das principais aplicações do Pandas é a limpeza e transformação de dados para análise.


````python
# Criando um DataFrame com dados "sujos" para demonstrar limpeza
import numpy as np

dados_sujos = {
    'Nome': ['josé','João', 'Ana', 'PEDRO', 'maria', 'Carlos', ''],
    'Idade': [80, 28, 24, 35, None, 45, 30],
    'Salario': [7000, 5000, 6000, None, 4500, 7000, 5500],
    'Cidade': ['Brasília', 'São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador', 'Curitiba', 'Fortaleza']
}

df_sujo = pd.DataFrame(dados_sujos)
print("DataFrame com dados 'sujos':")
df_sujo

````

````python
# Verificando dados faltantes
print("Dados faltantes por coluna:")
print(df_sujo.isnull().sum())
print()
````

````python
# Limpando os dados
df_limpo = df_sujo.copy()

# 1. Removendo linhas com nome vazio
df_limpo = df_limpo[df_limpo['Nome'] != '']

# 2. Padronizando nomes (primeira letra maiúscula)
df_limpo['Nome'] = df_limpo['Nome'].str.title()

# 3. Preenchendo valores faltantes
df_limpo['Idade'] = df_limpo['Idade'].fillna(df_limpo['Idade'].median())
df_limpo['Salario'] = df_limpo['Salario'].fillna(df_limpo['Salario'].mean())
print("DataFrame após limpeza:")
df_limpo
````

````python
# Criando nova coluna baseada em outra
df_limpo['Faixa_Etaria'] = df_limpo['Idade'].apply(lambda x: 'Jovem' if x < 30 else 'Adulto' if x < 50 else 'Maduro')
df_limpo['Salario_Categoria'] = pd.cut(df_limpo['Salario'], bins=[0, 5000, 6000, 10000], labels=['Baixo', 'Médio', 'Alto'])

print("DataFrame com novas colunas:")
df_limpo
````

## 📈 Operações de Agrupamento e Agregação

O `groupby` é uma das funcionalidades mais poderosas do Pandas para análise de dados.


````python
# Criando um DataFrame mais robusto para demonstrar groupby
np.random.seed(42)

dados_vendas = {
    'Vendedor': ['Ana', 'João', 'Pedro', 'Ana', 'João', 'Pedro', 'Ana', 'João', 'Pedro', 'Ana'] * 3,
    'Produto': ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Headset'] * 6,
    'Regiao': ['Norte', 'Sul', 'Nordeste', 'Centro-Oeste', 'Sudeste'] * 6,
    'Vendas': np.random.randint(1000, 10000, 30),
    'Quantidade': np.random.randint(1, 20, 30)
}

df_vendas_completo = pd.DataFrame(dados_vendas)
print("DataFrame de vendas:")
df_vendas_completo.head(10)
````

````python

# Agrupamento simples - vendas por vendedor
vendas_por_vendedor = df_vendas_completo.groupby('Vendedor')['Vendas'].sum()
print("Vendas totais por vendedor:")
vendas_por_vendedor
````

````python

# Múltiplas agregações
agregacoes = df_vendas_completo.groupby('Vendedor')['Vendas'].agg(['sum', 'mean', 'count'])
print("Múltiplas agregações por vendedor:")
agregacoes
````

````python

# Agrupamento por múltiplas colunas
vendas_por_vendedor_regiao = df_vendas_completo.groupby(['Vendedor', 'Regiao'])['Vendas'].sum()
print("Vendas por vendedor e região:")
vendas_por_vendedor_regiao
````

````python

# Pivot table
pivot_vendas = df_vendas_completo.pivot_table(
    values='Vendas', 
    index='Vendedor', 
    columns='Produto', 
    aggfunc='sum', 
    fill_value=0
)
print("Pivot table - Vendas por Vendedor x Produto:")
pivot_vendas
````

````python
# Estatísticas descritivas por grupo
print("Estatísticas descritivas por região:")
df_vendas_completo.groupby('Regiao')['Vendas'].describe()
````
