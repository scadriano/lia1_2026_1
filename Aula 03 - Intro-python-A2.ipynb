# 📈 2. Visualização de Dados com Matplotlib

## 🌸 Nosso Dataset: Iris
Vamos começar com um dos datasets mais famosos da ciência de dados: o **Iris Dataset**!

### 📋 Sobre o Dataset:
- **🌺 Espécies:** 3 tipos de flores íris (Setosa, Versicolor, Virginica)
- **📏 Medidas:** 4 características por flor (comprimento e largura de sépalas e pétalas)
- **📊 Amostras:** 150 observações (50 de cada espécie)
- **🎯 Objetivo:** Classificar espécies baseado nas medidas

> **💡 Por que o Iris?** É um dataset pequeno, limpo e perfeito para aprender visualização!


````python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configurações para gráficos mais bonitos
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# Carregando o dataset Iris
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
iris = pd.read_csv(url)

print("🌸 Dataset Iris carregado com sucesso!")
print("\n📊 Informações básicas do dataset:")
print(f"📏 Dimensões: {iris.shape}")
print(f"🌺 Espécies: {iris['species'].unique()}")
print(f"📋 Colunas: {list(iris.columns)}")

print("\n🔍 Primeiras 5 linhas:")
iris.head()
````

## 📊 Tipos de Gráficos com Matplotlib

### 🎯 O que vamos aprender:
1. **📈 Gráfico de Linhas** - Tendências ao longo de sequências
2. **📊 Histograma** - Distribuição de variáveis numéricas
3. **🔸 Gráfico de Dispersão** - Relação entre duas variáveis

---

### 📈 a) Gráfico de Linhas (`plt.plot`)
> **Quando usar:** Visualizar tendências de uma variável ao longo de uma sequência

**Ideal para:**
- 📅 Dados temporais (séries temporais)
- 📊 Comparar múltiplas variáveis
- 🔍 Identificar padrões e tendências


````python
# Gráfico de Linhas - Medidas das Sépalas
plt.figure(figsize=(12, 6))

# Plotando as duas séries de dados
plt.plot(iris['sepal_length'], label='Comprimento da Sépala', color='#2E86AB', linewidth=2)
plt.plot(iris['sepal_width'], label='Largura da Sépala', color='#A23B72', linewidth=2)

# Personalizando o gráfico
plt.title('Medidas das Sépalas das Flores de Íris', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Índice da Amostra', fontsize=12)
plt.ylabel('Medida (cm)', fontsize=12)
plt.legend(fontsize=11, loc='upper right')
plt.grid(True, alpha=0.3)

# Adicionando algumas anotações
plt.axhline(y=iris['sepal_length'].mean(), color='#2E86AB', linestyle='--', alpha=0.5, label='Média Comprimento')
plt.axhline(y=iris['sepal_width'].mean(), color='#A23B72', linestyle='--', alpha=0.5, label='Média Largura')

plt.tight_layout()
plt.show()

# Análise rápida
print("📊 Análise do Gráfico de Linhas:")
print(f"✓ Comprimento médio da sépala: {iris['sepal_length'].mean():.2f} cm")
print(f"✓ Largura média da sépala: {iris['sepal_width'].mean():.2f} cm")
print("✓ Observamos que o comprimento é consistentemente maior que a largura")
print("✓ As medidas variam de forma similar ao longo das amostras")
````

### 📊 b) Histograma (`plt.hist`)
> **Quando usar:** Entender a distribuição de uma única variável numérica

**Ideal para:**
- 📈 Visualizar distribuição de frequências
- 🔍 Identificar assimetrias e outliers
- 📊 Comparar distribuições entre grupos


````python
# Histograma - Distribuição do Comprimento das Pétalas
plt.figure(figsize=(10, 6))

# Criando o histograma
plt.hist(iris['petal_length'], bins=20, color='#52B788', alpha=0.7, edgecolor='black', linewidth=1.2)

# Personalizando o gráfico
plt.title('Distribuição do Comprimento das Pétalas', fontsize=16, fontweight='bold')
plt.xlabel('Comprimento (cm)', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.grid(True, alpha=1, axis='y')

# Adicionando linha da média
media_petal = iris['petal_length'].mean()
plt.axvline(x=media_petal, color='red', linestyle='--', linewidth=2, label=f'Média: {media_petal:.2f} cm')
plt.legend()

plt.tight_layout()
plt.show()

# Análise estatística detalhada
print("📊 Análise do Histograma:")
print(f"✓ Média: {iris['petal_length'].mean():.2f} cm")
print(f"✓ Mediana: {iris['petal_length'].median():.2f} cm")
print(f"✓ Desvio padrão: {iris['petal_length'].std():.2f} cm")
print(f"✓ Mínimo: {iris['petal_length'].min():.2f} cm")
print(f"✓ Máximo: {iris['petal_length'].max():.2f} cm")
print()
print("🔍 Observações:")
print("✓ O histograma mostra duas concentrações principais de dados")
print("✓ Isso sugere que existem grupos distintos de flores com pétalas de tamanhos diferentes")
print("✓ A distribuição é bimodal (dois picos), indicando possível separação por espécies")
````

### 🔸 c) Gráfico de Dispersão (`plt.scatter`)
> **Quando usar:** Observar a relação entre duas variáveis numéricas

**Ideal para:**
- 🔗 Identificar correlações
- 📈 Detectar padrões e tendências
- 🎯 Encontrar outliers
- 📊 Comparar grupos diferentes


````python
# Gráfico de Dispersão - Relação entre Comprimento e Largura da Pétala
plt.figure(figsize=(10, 6))

# Criando o gráfico de dispersão
plt.scatter(iris['petal_length'], iris['petal_width'], 
           color='#E63946', alpha=0.6, s=60, edgecolors='black', linewidth=0.5)

# Personalizando o gráfico
plt.title('Dispersão: Comprimento vs. Largura da Pétala', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Comprimento da Pétala (cm)', fontsize=12)
plt.ylabel('Largura da Pétala (cm)', fontsize=12)
plt.grid(True, alpha=0.3)

# Adicionando linha de tendência
z = np.polyfit(iris['petal_length'], iris['petal_width'], 1)
p = np.poly1d(z)
plt.plot(iris['petal_length'], p(iris['petal_length']), 
         color='blue', linestyle='--', linewidth=2, label='Linha de Tendência')

plt.legend()
plt.tight_layout()
plt.show()

# Calculando a correlação
correlacao = iris['petal_length'].corr(iris['petal_width'])

print("📊 Análise do Gráfico de Dispersão:")
print(f"✓ Correlação entre comprimento e largura da pétala: {correlacao:.3f}")
print(f"✓ Tipo de correlação: {'Positiva forte' if correlacao > 0.7 else 'Positiva moderada' if correlacao > 0.5 else 'Positiva fraca'}")
print()
print("🔍 Observações:")
print("✓ Existe uma forte correlação positiva entre comprimento e largura da pétala")
print("✓ Quanto maior o comprimento, maior tende a ser a largura")
print("✓ Os pontos se agrupam de forma quase linear, indicando relação consistente")
print("✓ Poucas observações fogem do padrão geral")
````

### 📊 a) Gráfico de Barras (`sns.barplot`)
> **Quando usar:** Comparar uma medida numérica entre diferentes categorias

**Características:**
- 📈 Calcula médias automaticamente
- 📊 Mostra barras de erro (intervalos de confiança)
- 🎯 Ideal para comparações entre grupos


````python
# Gráfico de Barras - Média do Comprimento da Sépala por Espécie
plt.figure(figsize=(10, 6))

# Criando o gráfico de barras com Seaborn
ax = sns.barplot(x='species', y='sepal_length', data=iris, 
                palette='viridis', errorbar=('ci', 95), capsize=0.1)

# Personalizando o gráfico
plt.title('Média do Comprimento da Sépala por Espécie', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Espécie', fontsize=12)
plt.ylabel('Comprimento da Sépala (cm)', fontsize=12)

# Adicionando valores nas barras
for i, v in enumerate(iris.groupby('species')['sepal_length'].mean()):
    ax.text(i, v + 0.05, f'{v:.2f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Análise estatística por espécie
print("📊 Análise Estatística por Espécie:")
print("=" * 50)
especies_stats = iris.groupby('species')['sepal_length'].agg(['mean', 'std', 'min', 'max'])
for especie, stats in especies_stats.iterrows():
    print(f"🌺 {especie.capitalize()}:")
    print(f"   ✓ Média: {stats['mean']:.2f} cm")
    print(f"   ✓ Desvio padrão: {stats['std']:.2f} cm")
    print(f"   ✓ Mínimo: {stats['min']:.2f} cm")
    print(f"   ✓ Máximo: {stats['max']:.2f} cm")
    print()

print("🔍 Observações:")
print("✓ Virginica tem o maior comprimento médio de sépala")
print("✓ Setosa tem o menor comprimento médio")
print("✓ As barras de erro mostram a variabilidade dos dados")
print("✓ Há diferenças significativas entre as espécies")
````

### 📦 b) Boxplot (`sns.boxplot`)
> **Quando usar:** Visualizar distribuição de dados, incluindo mediana, quartis e outliers

**Características:**
- 📊 **Mediana:** Linha central da caixa
- 📈 **Quartis:** Limites superior e inferior da caixa (Q1 e Q3)
- 🎯 **Outliers:** Pontos fora dos "bigodes"
- 📋 **Comparação:** Ideal para comparar distribuições entre grupos

**Interpretação:**
- 📦 **Caixa:** Contém 50% dos dados (Q1 a Q3)
- 📏 **Bigodes:** Estendem-se até 1.5 × IQR
- 🔸 **Pontos:** Outliers (valores extremos)


````python
# Boxplot - Distribuição do Comprimento da Pétala por Espécie
plt.figure(figsize=(10, 6))

# Criando o boxplot com Seaborn
ax = sns.boxplot(x='species', y='petal_length', data=iris, 
                palette='Set2', linewidth=2)

# Personalizando o gráfico
plt.title('📦 Distribuição do Comprimento da Pétala por Espécie', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Espécie', fontsize=12)
plt.ylabel('Comprimento da Pétala (cm)', fontsize=12)

# Adicionando pontos individuais (strip plot)
sns.stripplot(x='species', y='petal_length', data=iris, 
              color='black', alpha=0.3, size=4)

plt.tight_layout()
plt.show()

# Análise detalhada dos quartis
print("📊 Análise Detalhada dos Quartis:")
print("=" * 50)

for especie in iris['species'].unique():
    dados_especie = iris[iris['species'] == especie]['petal_length']
    q1 = dados_especie.quantile(0.25)
    q2 = dados_especie.quantile(0.50)  # Mediana
    q3 = dados_especie.quantile(0.75)
    iqr = q3 - q1
    
    print(f"🌺 {especie.capitalize()}:")
    print(f"   ✓ Q1 (25%): {q1:.2f} cm")
    print(f"   ✓ Q2 (Mediana): {q2:.2f} cm")
    print(f"   ✓ Q3 (75%): {q3:.2f} cm")
    print(f"   ✓ IQR: {iqr:.2f} cm")
    print(f"   ✓ Outliers: {len(dados_especie[(dados_especie < q1 - 1.5*iqr) | (dados_especie > q3 + 1.5*iqr)])}")
    print()

print("🔍 Observações:")
print("✓ Setosa tem pétalas muito menores que as outras espécies")
print("✓ Virginica tem maior variabilidade (caixa mais alta)")
print("✓ Há pouca sobreposição entre as espécies")
print("✓ Poucos outliers identificados")
````

### 🔄 c) Pairplot (`sns.pairplot`)
> **Quando usar:** Visualizar relacionamentos entre todas as variáveis do dataset

**Características:**
- 🔄 **Matriz completa:** Mostra todas as combinações de variáveis
- 📊 **Diagonal:** Histogramas das variáveis individuais
- 🔸 **Off-diagonal:** Gráficos de dispersão entre pares de variáveis
- 🎨 **Coloração:** Diferenciação por categorias (hue)

**Vantagens:**
- 🚀 **Visão geral:** Análise completa do dataset em um gráfico
- 🔍 **Detecção de padrões:** Identificação rápida de correlações
- 📈 **Comparação:** Diferenças entre grupos


````python
# Pairplot - Matriz de Relacionamentos
print("🔄 Gerando Pairplot... (pode demorar alguns segundos)")

# Criando o pairplot com coloração por espécie
g = sns.pairplot(iris, hue='species', 
                markers=["o", "s", "D"],
                palette='Set1',
                plot_kws={'alpha': 0.7, 's': 50},
                diag_kind='hist')

# Personalizando o título
g.fig.suptitle('Pair Plot das Variáveis do Dataset Iris', y=1.02, fontsize=16, fontweight='bold')

# Melhorando a legenda
handles = g._legend_data.values()
labels = g._legend_data.keys()
g._legend.remove()
g.fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 0.95), ncol=3, fontsize=12)

plt.tight_layout()
plt.show()

# Análise das correlações
print("\n📊 Matriz de Correlações:")
print("=" * 40)
correlacoes = iris.select_dtypes(include=[np.number]).corr()
print(correlacoes.round(3))

print("\n🔍 Principais Insights do Pairplot:")
print("✓ Diagonal: Histogramas mostram distribuições das variáveis por espécie")
print("✓ Petal_length vs Petal_width: Forte correlação positiva")
print("✓ Sepal_length vs Petal_length: Correlação moderada")
print("✓ Espécies são bem separáveis nas variáveis das pétalas")
print("✓ Setosa é claramente distinta das outras espécies")
print("✓ Virginica e Versicolor têm alguma sobreposição")

# Correlações mais fortes
print("\n🔗 Correlações mais fortes:")
for i in range(len(correlacoes.columns)):
    for j in range(i+1, len(correlacoes.columns)):
        corr = correlacoes.iloc[i, j]
        if abs(corr) > 0.5:
            print(f"✓ {correlacoes.columns[i]} vs {correlacoes.columns[j]}: {corr:.3f}")
````

# 🚢 4. Estudo de Caso: Análise do Dataset Titanic

## 🎯 Objetivo do Estudo
Vamos aplicar todas as técnicas aprendidas para analisar um dos datasets mais famosos da ciência de dados: **o desastre do Titanic**!

### 🔍 Perguntas que vamos responder:
1. **👥 O sexo influenciou na sobrevivência?**
2. **🏛️ A classe social teve impacto na sobrevivência?**
3. **🎂 Qual a relação entre idade e sobrevivência?**
4. **💰 O preço da passagem influenciou as chances de sobrevivência?**

### 📊 Sobre o Dataset:
- **📅 Data:** 15 de abril de 1912
- **👥 Passageiros:** 2,224 pessoas a bordo
- **💔 Sobreviventes:** Aproximadamente 710 pessoas
- **🔢 Variáveis:** Idade, sexo, classe, tarifa, etc.

> **🎯 Meta:** Descobrir padrões que influenciaram a sobrevivência através de visualizações!


## 📁 a) Carregando e Preparando os Dados

### 🧹 Etapas de Limpeza:
1. **📥 Carregar** o dataset
2. **🔍 Explorar** a estrutura dos dados
3. **🧹 Limpar** valores ausentes
4. **🗂️ Preparar** para análise


````python
# Carregando o dataset Titanic
print("🚢 Carregando dataset Titanic...")
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')

print("📊 Informações básicas do dataset:")
print(f"📏 Dimensões: {titanic.shape}")
print(f"📋 Colunas: {list(titanic.columns)}")

print("\n🔍 Primeiras 5 linhas:")
titanic.head()
````

````python

print("\n🧹 Verificando dados ausentes:")
dados_ausentes = titanic.isnull().sum()
print(dados_ausentes[dados_ausentes > 0])

print("\n🔧 Realizando limpeza dos dados...")

# Limpeza rápida baseada no material
# 1. Preencher idades faltantes com a média
idade_media = titanic['age'].mean()
titanic['age'] = titanic['age'].fillna(idade_media)

# 2. Remover a coluna 'deck' por ter muitos dados faltantes
if 'deck' in titanic.columns:
    titanic = titanic.drop('deck', axis=1)

# 3. Preencher embarked com o valor mais comum
if titanic['embarked'].isnull().sum() > 0:
    titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])

# 4. Remover linhas com tarifa ausente (se houver)
titanic = titanic.dropna(subset=['fare'])

print("✅ Dataset Titanic pronto para análise!")
print(f"📏 Dimensões finais: {titanic.shape}")
print(f"🧹 Dados ausentes restantes: {titanic.isnull().sum().sum()}")

# Estatísticas gerais de sobrevivência
taxa_sobrevivencia = titanic['survived'].mean()
print(f"\n📊 Taxa geral de sobrevivência: {taxa_sobrevivencia:.1%}")
print(f"👥 Total de passageiros: {len(titanic)}")
print(f"✅ Sobreviventes: {titanic['survived'].sum()}")
print(f"💔 Não sobreviventes: {len(titanic) - titanic['survived'].sum()}")
````

## 📊 b) Análise Exploratória com Gráficos

Vamos usar visualizações para responder nossas perguntas de pesquisa!

### 👥 Pergunta 1: O sexo influenciou na sobrevivência?
> **Hipótese:** A política de "mulheres e crianças primeiro" deve ter influenciado as taxas de sobrevivência


````python
# Pergunta 1: O sexo influenciou na sobrevivência?
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: Contagem de sobreviventes por sexo
sns.countplot(x='sex', hue='survived', data=titanic, palette='pastel', ax=ax1)
ax1.set_title('Sobrevivência por Sexo', fontsize=14, fontweight='bold')
ax1.set_ylabel('Número de Passageiros')
ax1.legend(title='Sobreviveu', labels=['Não', 'Sim'])

# Gráfico 2: Taxa de sobrevivência por sexo
sobrevivencia_sexo = titanic.groupby('sex')['survived'].mean()
ax2.bar(sobrevivencia_sexo.index, sobrevivencia_sexo.values, 
        color=['#ff9999', '#66b3ff'], alpha=0.7)
ax2.set_title('Taxa de Sobrevivência por Sexo', fontsize=14, fontweight='bold')
ax2.set_ylabel('Taxa de Sobrevivência')
ax2.set_ylim(0, 1)

# Adicionando valores nas barras
for i, v in enumerate(sobrevivencia_sexo.values):
    ax2.text(i, v + 0.02, f'{v:.1%}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Análise estatística
print("📊 Análise Estatística - Sobrevivência por Sexo:")
print("=" * 50)
tabela_sexo = pd.crosstab(titanic['sex'], titanic['survived'], margins=True)
print(tabela_sexo)

print(f"\n🔍 Insights:")
print(f"✓ Taxa de sobrevivência feminina: {sobrevivencia_sexo['female']:.1%}")
print(f"✓ Taxa de sobrevivência masculina: {sobrevivencia_sexo['male']:.1%}")
print(f"✓ Diferença: {sobrevivencia_sexo['female'] - sobrevivencia_sexo['male']:.1%}")

print(f"\n📋 Conclusões:")
print("✅ A política de 'mulheres e crianças primeiro' foi claramente aplicada")
print("✅ Mulheres tiveram uma taxa de sobrevivência muito superior aos homens")
print("✅ O sexo foi um fator determinante para a sobrevivência")
````

### 🏛️ Pergunta 2: A classe social teve impacto na sobrevivência?
> **Hipótese:** Passageiros da primeira classe tiveram melhor acesso aos botes salva-vidas


````python
# Pergunta 2: A classe social teve impacto na sobrevivência?
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: Contagem de sobreviventes por classe
sns.countplot(x='class', hue='survived', data=titanic, 
              palette='viridis', order=['First', 'Second', 'Third'], ax=ax1)
ax1.set_title('Sobrevivência por Classe de Passageiro', fontsize=14, fontweight='bold')
ax1.set_ylabel('Número de Passageiros')
ax1.legend(title='Sobreviveu', labels=['Não', 'Sim'])

# Gráfico 2: Taxa de sobrevivência por classe
sobrevivencia_classe = titanic.groupby('class')['survived'].mean()
ordem_classes = ['First', 'Second', 'Third']
cores_classes = ['#d4af37', '#c0c0c0', '#cd7f32']  # Dourado, Prata, Bronze

ax2.bar(range(len(ordem_classes)), 
        [sobrevivencia_classe[classe] for classe in ordem_classes],
        color=cores_classes, alpha=0.7)
ax2.set_title('Taxa de Sobrevivência por Classe', fontsize=14, fontweight='bold')
ax2.set_ylabel('Taxa de Sobrevivência')
ax2.set_xticks(range(len(ordem_classes)))
ax2.set_xticklabels(ordem_classes)
ax2.set_ylim(0, 1)

# Adicionando valores nas barras
for i, classe in enumerate(ordem_classes):
    valor = sobrevivencia_classe[classe]
    ax2.text(i, valor + 0.02, f'{valor:.1%}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Análise estatística detalhada
print("📊 Análise Estatística - Sobrevivência por Classe:")
print("=" * 50)
tabela_classe = pd.crosstab(titanic['class'], titanic['survived'], margins=True)
print(tabela_classe)

print(f"\n🔍 Insights por Classe:")
for classe in ordem_classes:
    taxa = sobrevivencia_classe[classe]
    total = len(titanic[titanic['class'] == classe])
    sobreviventes = titanic[(titanic['class'] == classe) & (titanic['survived'] == 1)]
    print(f"✓ {classe}: {taxa:.1%} ({len(sobreviventes)}/{total} passageiros)")

print(f"\n📋 Conclusões:")
print("✅ Clara hierarquia social nas taxas de sobrevivência")
print("✅ Primeira classe teve quase 2x mais chances que a terceira")
print("✅ A classe social foi um fator determinante para sobrevivência")
print("✅ Recursos e localização no navio influenciaram o acesso aos botes")
````

### 🎂 Pergunta 3: Qual a relação entre idade e sobrevivência?
> **Hipótese:** Crianças tiveram prioridade nos botes salva-vidas ("mulheres e crianças primeiro")


````python
# Pergunta 3: Qual a relação entre idade e sobrevivência?
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: Violinplot - Distribuição de idades por sobrevivência
sns.violinplot(x='survived', y='age', data=titanic, palette='muted', ax=ax1)
ax1.set_title('Distribuição de Idades: Sobreviventes vs Não Sobreviventes', fontsize=14, fontweight='bold')
ax1.set_xticklabels(['Não Sobreviveu', 'Sobreviveu'])
ax1.set_ylabel('Idade (anos)')

# Gráfico 2: Histograma comparativo de idades
ax2.hist(titanic[titanic['survived'] == 0]['age'], bins=20, alpha=0.7, 
         label='Não Sobreviveu', color='red', density=True)
ax2.hist(titanic[titanic['survived'] == 1]['age'], bins=20, alpha=0.7, 
         label='Sobreviveu', color='green', density=True)
ax2.set_title('Distribuição de Idades por Sobrevivência', fontsize=14, fontweight='bold')
ax2.set_xlabel('Idade (anos)')
ax2.set_ylabel('Densidade')
ax2.legend()

plt.tight_layout()
plt.show()

# Análise por faixas etárias
print("📊 Análise por Faixas Etárias:")
print("=" * 40)

# Criando categorias de idade
titanic['faixa_etaria'] = pd.cut(titanic['age'], 
                                bins=[0, 12, 25, 60, 100],
                                labels=['Criança (0-12)', 'Jovem (13-25)', 'Adulto (26-60)', 'Idoso (60+)'])

# Calculando taxa de sobrevivência por faixa etária
sobrevivencia_idade = titanic.groupby('faixa_etaria')['survived'].agg(['mean', 'count'])
sobrevivencia_idade.columns = ['Taxa_Sobrevivencia', 'Total_Passageiros']

print(sobrevivencia_idade)

# Estatísticas descritivas
print(f"\n📈 Estatísticas de Idade:")
print(f"✓ Idade média dos sobreviventes: {titanic[titanic['survived'] == 1]['age'].mean():.1f} anos")
print(f"✓ Idade média dos não sobreviventes: {titanic[titanic['survived'] == 0]['age'].mean():.1f} anos")
print(f"✓ Diferença: {titanic[titanic['survived'] == 1]['age'].mean() - titanic[titanic['survived'] == 0]['age'].mean():.1f} anos")

print(f"\n🔍 Insights por Faixa Etária:")
for faixa in sobrevivencia_idade.index:
    taxa = sobrevivencia_idade.loc[faixa, 'Taxa_Sobrevivencia']
    total = sobrevivencia_idade.loc[faixa, 'Total_Passageiros']
    print(f"✓ {faixa}: {taxa:.1%} ({total} passageiros)")

print(f"\n📋 Conclusões:")
print("✅ Crianças tiveram a maior taxa de sobrevivência")
print("✅ A política 'mulheres e crianças primeiro' foi aplicada")
print("✅ Idade foi um fator importante, mas não o mais determinante")
print("✅ A variabilidade dentro das faixas etárias é significativa")
````

### 💰 Pergunta 4: O preço da passagem influenciou as chances de sobrevivência?
> **Hipótese:** Passageiros que pagaram mais caro tiveram melhor acesso aos recursos de salvamento


````python
# Pergunta 4: O preço da passagem influenciou as chances de sobrevivência?
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gráfico 1: Boxplot do preço da passagem por sobrevivência
sns.boxplot(x='survived', y='fare', data=titanic, palette='coolwarm', ax=ax1)
ax1.set_title('Preço da Passagem: Sobreviventes vs Não Sobreviventes', fontsize=14, fontweight='bold')
ax1.set_xticklabels(['Não Sobreviveu', 'Sobreviveu'])
ax1.set_ylabel('Preço da Passagem (£)')

# Gráfico 2: Análise por faixas de preço
# Criando categorias de preço
titanic['faixa_preco'] = pd.cut(titanic['fare'], 
                               bins=[0, 10, 30, 100, 1000],
                               labels=['Baixo (0-10)', 'Médio (10-30)', 'Alto (30-100)', 'Premium (100+)'])

sobrevivencia_preco = titanic.groupby('faixa_preco')['survived'].mean()
ax2.bar(range(len(sobrevivencia_preco)), sobrevivencia_preco.values, 
        color=['#8B0000', '#FF6B6B', '#4ECDC4', '#45B7D1'], alpha=0.7)
ax2.set_title('Taxa de Sobrevivência por Faixa de Preço', fontsize=14, fontweight='bold')
ax2.set_ylabel('Taxa de Sobrevivência')
ax2.set_xticks(range(len(sobrevivencia_preco)))
ax2.set_xticklabels(sobrevivencia_preco.index, rotation=45)
ax2.set_ylim(0, 1)

# Adicionando valores nas barras
for i, v in enumerate(sobrevivencia_preco.values):
    ax2.text(i, v + 0.02, f'{v:.1%}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Análise estatística
print("📊 Análise Estatística - Preço da Passagem:")
print("=" * 50)

print(f"💰 Preço médio dos sobreviventes: £{titanic[titanic['survived'] == 1]['fare'].mean():.2f}")
print(f"💰 Preço médio dos não sobreviventes: £{titanic[titanic['survived'] == 0]['fare'].mean():.2f}")
print(f"💰 Diferença: £{titanic[titanic['survived'] == 1]['fare'].mean() - titanic[titanic['survived'] == 0]['fare'].mean():.2f}")

print(f"\n🔍 Insights por Faixa de Preço:")
faixas_preco = titanic.groupby('faixa_preco')['survived'].agg(['mean', 'count'])
for faixa in faixas_preco.index:
    taxa = faixas_preco.loc[faixa, 'mean']
    total = faixas_preco.loc[faixa, 'count']
    print(f"✓ {faixa}: {taxa:.1%} ({total} passageiros)")

# Correlação entre preço e sobrevivência
correlacao_preco = titanic['fare'].corr(titanic['survived'])
print(f"\n📈 Correlação preço-sobrevivência: {correlacao_preco:.3f}")

print(f"\n📋 Conclusões:")
print("✅ Passageiros que pagaram mais caro tiveram maior taxa de sobrevivência")
print("✅ Há uma clara relação entre poder aquisitivo e chances de salvamento")
print("✅ A faixa premium teve quase 2x mais chances que a faixa baixa")
print("✅ O preço da passagem reflete a classe social e localização no navio")
````

### 🔄 Análise Combinada: Interação entre Variáveis
> **Objetivo:** Entender como múltiplos fatores interagem para influenciar a sobrevivência


````python
# Análise Combinada: Interação entre Variáveis
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Gráfico 1: Sobrevivência por Sexo e Classe
sobrevivencia_sexo_classe = titanic.groupby(['sex', 'class'])['survived'].mean().unstack()
sobrevivencia_sexo_classe.plot(kind='bar', ax=ax1, color=['#d4af37', '#c0c0c0', '#cd7f32'])
ax1.set_title('Taxa de Sobrevivência por Sexo e Classe', fontsize=14, fontweight='bold')
ax1.set_ylabel('Taxa de Sobrevivência')
ax1.set_xlabel('Sexo')
ax1.legend(title='Classe')
ax1.tick_params(axis='x', rotation=0)

# Gráfico 2: Heatmap de Sobrevivência
pivot_table = titanic.pivot_table(values='survived', index='sex', columns='class', aggfunc='mean')
sns.heatmap(pivot_table, annot=True, fmt='.3f', cmap='RdYlGn', ax=ax2)
ax2.set_title('Heatmap: Taxa de Sobrevivência por Sexo e Classe', fontsize=14, fontweight='bold')

# Gráfico 3: Sobrevivência por Idade e Sexo
sns.boxplot(x='sex', y='age', hue='survived', data=titanic, ax=ax3)
ax3.set_title('Idade por Sexo e Sobrevivência', fontsize=14, fontweight='bold')
ax3.legend(title='Sobreviveu', labels=['Não', 'Sim'])

# Gráfico 4: Scatter plot - Idade vs Preço colorido por sobrevivência
scatter = ax4.scatter(titanic['age'], titanic['fare'], 
                     c=titanic['survived'], cmap='coolwarm', alpha=0.6)
ax4.set_title('Idade vs Preço da Passagem (colorido por sobrevivência)', fontsize=14, fontweight='bold')
ax4.set_xlabel('Idade (anos)')
ax4.set_ylabel('Preço da Passagem (£)')
plt.colorbar(scatter, ax=ax4, label='Sobreviveu')

plt.tight_layout()
plt.show()

# Análise estatística da interação
print("📊 Análise de Interação entre Variáveis:")
print("=" * 50)

print("🔍 Taxa de Sobrevivência por Sexo e Classe:")
for sexo in ['male', 'female']:
    print(f"\n👤 {sexo.capitalize()}:")
    for classe in ['First', 'Second', 'Third']:
        subset = titanic[(titanic['sex'] == sexo) & (titanic['class'] == classe)]
        if len(subset) > 0:
            taxa = subset['survived'].mean()
            total = len(subset)
            print(f"   ✓ {classe}: {taxa:.1%} ({total} passageiros)")

# Casos extremos
print(f"\n🎯 Casos Extremos:")
print(f"✅ Melhor chance: Mulheres da 1ª classe ({sobrevivencia_sexo_classe.loc['female', 'First']:.1%})")
print(f"❌ Pior chance: Homens da 3ª classe ({sobrevivencia_sexo_classe.loc['male', 'Third']:.1%})")

# Correlações
print(f"\n📈 Correlações com Sobrevivência:")
correlacoes = titanic[['survived', 'age', 'fare']].corr()['survived'].sort_values(ascending=False)
for var, corr in correlacoes.items():
    if var != 'survived':
        print(f"✓ {var}: {corr:.3f}")

print(f"\n📋 Insights da Análise Combinada:")
print("✅ Sexo foi o fator mais determinante para sobrevivência")
print("✅ Classe social ampliou as diferenças entre homens e mulheres")
print("✅ Mulheres da 1ª classe tiveram quase 100% de sobrevivência")
print("✅ Homens da 3ª classe tiveram as menores chances")
print("✅ Idade teve impacto menor comparado a sexo e classe")
print("✅ Preço da passagem correlaciona com classe social")
````
